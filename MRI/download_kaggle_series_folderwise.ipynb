```python
import shutil
# import os
from IPython.display import FileLink, display

# Set the Series UID
series_uid = "1.2.826.0.1.3680043.8.498.10037266473301611864455091971206084528"

# Source and target paths
src = f"/kaggle/input/rsna-intracranial-aneurysm-detection/series/{series_uid}"
dst = f"/kaggle/working/{series_uid}"

# Copy the folder to /kaggle/working so we can zip it
shutil.copytree(src, dst)

# Copy train.csv to /kaggle/working
shutil.copy(
    "/kaggle/input/rsna-intracranial-aneurysm-detection/train.csv",
    "/kaggle/working/train.csv"
)
```

```python
# Create ZIP archive of the copied folder
shutil.make_archive(series_uid, 'zip', root_dir=f"/kaggle/working", base_dir=series_uid)
```

```python
# Display download link
display(FileLink(f"{series_uid}.zip"))
display(FileLink("train.csv"))
```
